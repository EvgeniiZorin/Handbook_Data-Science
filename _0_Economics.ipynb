# Price indices and inflation

**Inflation** (by default, relates to *price inflation*) - general increase in price of goods and services.
- Inflation rate: the pace at which the overall price level is increasing; 
- A little inflation is good: 1-3% per year
- More than that - concerning

Inflation is measured by **CPI** (consumer price index) - measures the monthly change in price for a figurative basket of goods and services.
- CPI can be CPI-U (CPI for all urban consumers) which represents 93% of the U.S. population not living in remote rural areas, and CPI-W.
- CPI is an index that calculates the cost of a market basket of goods purchased by a typical family that lives in an urban area; the purpose of the CPI is to track changes in the cost of living over time.
- market basket is the combination of goods that are used to calculate a price index; the goods stay the same from year to year.

You calculate CPI like this:
- Start at some chosen year
- Calculate the market basket price for each month of the dataset
- For the earliest period of your data (e.g. the very first month), you set it as CPI = 100
- For the next month, you can calculate CPI:

$$
\text{CPI} = \cfrac{\text{Current Month Cost}}{\text{Base Month Cost}} * 100
$$

- Then, calculate like this for every month
- Finally, you can calculate inflation rate month to month: 

$$
\text{Inflation Rate} = \cfrac{\text{Current Month CPI} - \text{Previous Month CPI}}{\text{Previous Month CPI}} * 100
$$

> The same calculations can be done on a different level of aggregation, e.g. yearly
