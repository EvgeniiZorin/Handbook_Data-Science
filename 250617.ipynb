# Import libraries and init variables

````python
from datetime import datetime
````

# Start

````python
input1 = [
  {"user": "alice", "action": "login", "timestamp": "2023-01-01T12:00:00"},
  {"user": "alice", "action": "upload", "timestamp": "2023-01-01T12:05:00"},
  {"user": "bob", "action": "login", "timestamp": "2023-01-01T12:10:00"}
]
input1
````

````python
def summarize_user_activity(data: list[dict]) -> dict:
    """
    A brief explanation here.

    Parameters:
        data: list[dict] - explanation 
    """
    # Create empty dict output
    dict1 = dict()
    # Count the number of actions per user
    for i in data:
        name = i['user']
        if name not in dict1:
            dict1[name] = 1
        else:
            dict1[name] += 1
    # Process the datetime
    dict2 = dict()
    for i in data:
        name = i['user']
        date = i['timestamp']
        if name not in dict2:
            dict2[name] = date
        else:
            dict2[name] = max(dict2[name], date)
    # final output
    final_dict = dict()
    for i in dict1.keys():
        final_dict[i] = {'actions': dict1[i], 'last_seen': dict2[i]}
    return final_dict

summarize_user_activity(input1)
````
