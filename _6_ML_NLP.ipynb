## Main tasks

- Text Classification:
  - E.g. Spam Filtering, tagging customer feedback into categories, routing support tickets according to their language;
  - **Sentiment Analysis**: identifying the polarity of text
    - e.g. "I love this movie!" has a positive polarity / sentiment
    - Sentiment Analysis can be used by e.g. a car company to analyse Twitter posts to determine whether people like the new feature of a car or not 
- Text-to-text Generation
  - **Machine Translation**: translating from one language to another
  - Text Generation: creating text that apperas indistinguishable from human-written text
  - Text Summarisation
  - Text Simplification
  - Lexical Normalisation: translating / transforming a non-standard text to a standard register
  - Paraphrase Generation
- Data-to-Data convertion:
  - Text-to-Speech
  - **Speech-to-Text** (Speech Recognition): decoding human voice and turning it into text
  - Text-to-Image
  - Data-to-Text: producing text from non-linguistic input, such as databases of records, spreadsheets, and expert system knowledge bases
- Knowledge bases, entities and relations
  - Relation extraction: extracting semantic relationships from a text. 
  - Relation prediction
  - **Named Entity Recognition (NER)**: tagging entities in text with their corresponding type, typically in BIO notation; e.g. names of individuals, organisations, locations, expressions of time, quantities, medical codes, monetary values and percentages, etc.
    - E.g. from the text 'Harrison Ford was in Star Wars.' we can extract named entities "Harrison Ford" and "Star Wars"
  - Entity Linking: recognising and disambiguating named entities to a knowledge base
- Topics and Keywords
  - Topic Modeling: identifying abstract topics underlying a collection of documents. Usually an unsupervised task discovering abstract topics. 
  - Keyword Extraction: identifying the most relevant terms to describe the subject of a document
- Information retrieval: finds the documents that are most relevant to a query

# History

- <u>1900s</u>
  - Andrey Markov, 1913 - created the first statistical language model, based on the Markov chain
    - Analysed the novel "Evgenii Onegin" by stripping out spaces and punctuation, then classifying each character as a vowel or consonant. He measures how often one followed the other, building a probabilistic system that could generate sequences of text based on prior context
    - [Read more here](https://spectrum.ieee.org/andrey-markov-and-claude-shannon-built-the-first-language-generation-models)
- <u>Early Rule-based systems (1950s-1980s)</u>
  - NLP started with handcrafted, rule-based systems for language understanding; 
  - Challenges: labor-intensive, brittle, unable to scale to complex language variations. 
  - Examples:
    - ELIZA (1966): 
      - Joseph Weizenbaum created ELIZA, one of the first chatbots capable of engaging in conversations with humans and making them believe the software had human-like emotions;
      - The dialogue and response options were hard-coded / canned 
    - SHRDLU (1968): 
      - Terry Winograd created a program called SHRDLU - the first multimodal AI that could manipulate and reason out of a world of blocks - BLOCKS world - according to instructions from a user;
      - Seemingly, there was dialog (in plain English) between the user and the computer program via instructions (via teletype), about a small world of objects, the BLOCKS world, shown on an early computer screen. The user interacts with the application by moving objects, naming collections, and asking questions about the condition of a condensed blocks world, a virtual box filled with various blocks:
      - 
- <u>Statistical Methods and N-grams (1990s)</u>
  - Shift from rule-based to statistical models using data. 
  - Language modeled as probabilistic sequences, e.g. N-gram models estimating probability; 
  - Used for part-of-speech tagging, speech recognition
  - Challenges: limited context window, data sparsity issues
- <u>ML (2000s)</u>
  - Supervised ML algorithms with engineered features
  - Models like Maximum Entropy, SVM trained on linguistic features (e.g. word shape, prefixes)
  - Improved tasks like NER, parsing
  - Challenges: feature engineering was costly and task-specific
- <u>Word Embeddings (2013)</u>
  - Dense vector representations of words capturing semantic similarity
  - Models like Word2Vec, GloVe learn word vectors by predicting context words, embedding words into continuous vector spaces; 
  - Enabled transfer learning and better generalisation by capturing semantic relationships
- <u>Recurrent Neural Networks and LSTMs (mid-2010s)</u>
  - RNN-based models, like LSTM
  - Sequence models that can capture longer context
  - RNNs process sequences sequentially; LSTMs add gating mechanisms to combat vanishing gradients, enabling learning of long-range dependencies
  - Used in machine translation, speech recognition
  - Challenges: training is sequential and slow, difficulty with very long dependencies
- <u>Attention Mechanism (2015)</u>
  - Allow models to selectively focus on parts of input sequences
  - Attention computes weighted sums of encoder states, enabling dynamic context modeling
  - Boosted performance on translation and text summarisation
- <u>Transformer (2017)</u>
  - Entirely attention-based architecture, discarding recurrence and convolutions
  - Uses self-attention layers to model dependencies in parallel, greatly improving training speed and scalability
  - Paper: "Attention is All You Need"
  - State-of-the-art in translation and many NLP tasks
- <u>Pretrained language models (2018-present)</u>
  - Transformers that combined self-attention with transfer learning
  - Train LLM on massive text corpora, then fine-tune on downstream tasks
  - Models like BERT (use masked language modeling), GPT series (use autoregressive language modeling)
  - Huge performance gains on tasks like question answering, sentiment analysis, NER
  - Models with hundreds of millions to billions of parameters trained on diverse data
- <u>LLM and Few-Shot Learning (2020+)</u>
  - GPT-3 and similar models show strong capabilities in zero-shot and few-shot learning
  - Very large Transformer models trained on massive corpora; prompt-based learning enables new tasks with minimal fine-tuning
  - Opens new frontiers for NLP as a general-purpose language understanding and generation tool

## Tokenisation

Tokenization - splitting the raw text into individual units (tokens). 

- Sentence tokenisation - split the text into sentences; 
- Word tokenisation - split the text into words; 
- Subword tokenisation - break down words into smaller subword units, e.g. "walking" -> "walk" and "ing";
- Character tokenisation: into characters, e.g. "walking" -> "w", "a", "l", "k", "i", "n", "g"

Some terms:
- Vocabulary: list of learned words / tokens
- Out of Vocabulary (OOV) words: words / tokens that were not seen at training but were encountered in testing. 

> Note: in some cases, you could treat some punctuation as tokens, e.g. so that two sentences - "I have cats." and "I have cats?" - have different encodings and thus interpretations / classifications.

### Stemming

- Process of eliminating affixes (suffixes, prefixes, infixes, circumfixes) from words to obtains their stems / base words. 
- It simply removes the ends of the words - a very crude process.

It is a more simple process compared to lemmatisation, because stem may not be an actual word, e.g.:
- `caring > car`, 
- `better > better`,
- `replacement > replac`
- `universal, university, universe > univers`
- `mice > mice`


````python
"""
There are multiple algorithms for stemming, e.g. Porter Stemmer in NLTK
"""
from nltk.stem import PorterStemmer
porter = PorterStemmer()

for i in ['walking', 'walked', 'walks', 'ran', 'run', 'running']: 
    stemmed = porter.stem(i)
    print(f"{i} > {stemmed}")

print('\n')
for i in ['caring', 'car', 'cars', 'better', 'replacement', 'universal', 'university', 'universe', 'mice', 'unnecessary', 'berry']:
    stemmed = porter.stem(i)
    print(f"{i} > {stemmed}")
````

### Lemmatisation

- Lemmatisation is related to stemming, differing in that lemmatization is able to capture canonical forms based on a word's lemma as well as context. 
- It actually uses grammatical language rules. 

E.g.:
- `caring > care`
- `likes > like`, 
- `better > good`, 
- `worse > bad`, 
- `study, studying, studies > study`
- `mice > mouse`


````python
"""
lemmatisation implementations appear in NLTK, spaCy, etc.
"""
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet 
# nltk.download('wordnet')
# nltk.download('averaged_perceptron_tagger_eng') # needed for Part-Of-Speech tagging

lemmatizer = WordNetLemmatizer()
for i in ['mouse', 'mice', 'acne', 'acnes']:
    lemma = lemmatizer.lemmatize(i)
    print(f"{i} > {lemma}")
````

## One-hot encoding

Assigning a numerical integer index for each unique word, based on vector representing all possible words. 

If $n$ is the size of the vocabulary, then each word is represented by a $n$-sized vector with all zeros except for the index representing this word. E.g. $n=3$:

Vocabulary, $n=3$:
- Apple: `[1,0,0]`
- Banana: `[0,1,0]`
- Orange: `[0,0,1]`

Word "apple": `[1,0,0]`

One sentence consisting of $k$ words can then be repsented as a 2D vector of shape $(l,n)$, where $l$ is a pre-specified maximum length of a sentence; therefore, if $k$ < $l$, we'll just pad the missing words in the rest of the sentence with zero-filled arrays. Let's say that $k=4$ and $l=5$:

Sentence "Apple apple apple banana":

```py
sentence_1 = [

    [0,0,0]
]
```

## Bag-of-Words

One-hot encoding can be used a little differently in count-based approaches, such as **Bag-of-Words (BOW)** approach, where we look at the histogram of the words within the text, i.e. considering each word count as a feature. 

For example, if we consider two sentences - "This is an apple", "This is a banana is banana": 
- The vocabulary will be `['this','is','an','apple','a','banana']`
- Now we can encode the sentences as two one-hot encoded vectors - `[1,1,1,1,0,0]` and `[1,2,0,0,1,2]`.
- A sentence "apple apple apple" will be encoded like this: `[0,0,0,3,0,0]`

<center><img src="Media/bag-of-words.png" width="550px"></center>

 **Advantages**: 
- The algorithm is simple, straightforward;
- Involves simple preprocessing procedures: tokenisation, normalisation, vocabulary generation, and vectorisation; 
- Robust to noise;

 **Disadvantages**:
- Loss of information: 
  - Does not conserve information about word proximity / closeness in meaning / relationship; in one-hot encoded vocabulary, every two words are the same distance (two size 1 steps) away from each other. This is not perfect, as 'banana' is closer in meaning to 'plantane' than 'king'. IOW, leads to loss of semantic meaning;
  - Loses word order;
- Cannot deal with unknown words - words that weren't seen in training;
- Very space inefficient: for each word, we have a vector that is both too big and sparse;
- Doesn't perform count normalisation, so longer sentences could have much, much higher counts of words, which could influence the decision (only based on the length of the sentence).

In theory, we could also normalise the counts in the vectors, e.g. by converting counts to probabilities, but currently it's not implemented because it's a very simple method.


### Count Vectoriser

This is simply an implementation of the Bag-of-Words approach:
- V: the number of unique words in the training corpus
- Each document in the text corpus is converted into a vector of size V, with counts per each word contained in the vector component

For example, if the vocabulary is `['and' 'document' 'first' 'is' 'one' 'second' 'the' 'third' 'this']`, then the sentence "This document is the second document." is converted into the following vector: `[[0 2 0 1 0 1 1 0 1]]`


````python
# Corpus for vocabulary training
corpus = [
    'This,; is the first document.',
    'This document is the second document.',
    'And this,,, is the third one.',
    'Is this the first document?',
]
# Sentences to process
list_of_docs = [
    'This is the document!',
    'That document is second document.'
]
````

````python
"""
Manual implementation of the Count Vectoriser (Bag-of-Words).
"""
import re

def text_normalisation(x:str):
    """
    Given an input string, outputs a normalised string
    which is lowercase, has punctuation and whitespaces removed
    """
    sentence = x
    ### lowercase
    sentence = sentence.lower()
    ### remove punctuation
    for j in ',.;?!':
        sentence = sentence.replace(j, ' ')
    ### remove whitespaces
    sentence = sentence.strip()
    sentence = re.sub(' +', ' ', sentence)
    return sentence

def fit_vocabulary(corpus:list):
    """
    given a corpus (a list of denormalised strings), outputs a vocabulary
    """
    ### text normalisation
    corpus_proc = []
    for i in corpus:
        sentence_norm = text_normalisation(i)
        corpus_proc.append(sentence_norm)
    ### first, get all unique words in the corpus and create a mapping / list, where index of each word represents its index in the transformed vector
    vocabulary_list = []
    for i in corpus_proc:
        for j in i.split(' '):
            if j not in vocabulary_list:
                vocabulary_list.append(j)
    vocabulary_list = sorted(vocabulary_list)
    return vocabulary_list

vocabulary = fit_vocabulary(corpus)
print(f"Vocabulary: {vocabulary}\n")

def transform_text(VOCABULARY:list, 
                   LIST_OF_DOCS:list):
    """
    vectorise a list of text strings
    """
    vectorised_collection = []
    for i in LIST_OF_DOCS:
        vectorised_doc = []
        # first, normalise the text string
        text_proc = text_normalisation(i)
        for j in VOCABULARY:
            
            count = len(re.findall(fr'(?:^|\W){j}(?:$|\W)', text_proc))
            vectorised_doc.append(count)
        vectorised_collection.append(vectorised_doc)
    return vectorised_collection

list_of_docs_vectorised = transform_text(vocabulary, list_of_docs)
print(f"""Original sentence: {list_of_docs}

Vectorised sentence: {list_of_docs_vectorised}

Mapping of the vectorised sentence:
""")
for index1,i in enumerate(list_of_docs_vectorised):
    print(f"Sentence: {list_of_docs[index1]}")
    for index2,j in enumerate(i):
        print(" > ", vocabulary[index2], j)
````

## (Bag of) N-gram

Another approach is **N-gram**. In this approach, we count groups of N words that are spatially neighbouring in the text. 


- Unigrams (1-gram): considers single words individually. This is basically the standard bag-of-words
- Bigrams (2-gram): two-word groups, e.g. "this is", "is an", "an apple", etc
- Trigrams (3-gram): 3 words
- N-grams (N-grams): N adjacent words

<center><img src="Media/n-grams.png" width="550px"></center>

Advantages:
- Capture more context (word order) than a simple BoW algorithm
- Can be used for predicting the next word in a sentence - chatbots, text summarisation systems, content generation tools
- Can be used to identify common phrases or expressions, which can be useful in sentiment analysis and document categorisation

Disadvantages:
- Very space inefficient: for each N-gram, we have a vector that is both too big and sparse; therefore, this can lead to high dimensionality, increased computational complexity, and high sparsity; 


````python
import nltk
from nltk.util import ngrams

text = "This is a sample sentence for generating n-grams."

### Tokenise the text into words
tokens = nltk.word_tokenize(text)
### Generate n-grams
n = 3 ### trigrams
ngrams_output = list(ngrams(tokens, n))
print(f'Generated n-grams: \n{ngrams_output}')
````

## TF-IDF

TF-IDF (Term Frequency - Inverse Document Frequency): statistical method to capture the relevance of words with respect to the corpus of text. Is a numerical statistic used in natural language processing and information retrieval to evaluate the importance of a term (word) within a document relative to a collection of documents (corpus).


**Term Frequency**: measures the frequency of appearance of a term in a document:

$$TF(t, d) = \cfrac{\text{Number of times a term $t$ appears in the document $d$}}{\text{Total number of terms in the document $d$}}$$

**Inverse Document Frequency**: measures how unique or rare a term is across a list of documents:

$$IDF(t, D) = log \left( \cfrac{\text{total number of documents in the corpus D}}{\text{Number of documents containing term t}} \right)$$

**TF-IDF**: The TF-IDF score for a term $t$ in a document $d$ is calculated by multiplying the term frequency (TF) of $t$ in $d$ by the inverse document frequency (IDF) of t across the entire corpus.

$$\text{TF-IDF}(t,d,D) = TF(t,d)  IDF(t,D)$$

$$t, \text{term}; d, \text{document}; D, \text{corpus}$$


An excellent example from a Medium post. 

Let's consider the corpus consisting of three documents:

```txt
Sentence 1: good boy
Sentence 2: good girl
Sentence 3: boy girl good
```

First let's calculate word frequency:
- `good`: 3
- `boy`: 2
- `girl`: 2

TF matrix for each sentence:

| | Sentence 1 | Sentence 2 | Sentence 3 |
| - | - | - | - |
| good | 0.5 | 0.5 | 1/3 |
| boy | 0.5 | 0 | 1/3 |
| girl | 0 | 0.5 | 1/3 |

IDF for the entire corpus:

| Word | IDF |
| - | - |
| good | log(3/3) = 0 |
| boy | log(3/2) = 0.176 |
| girl | log(3/2) = 0.176 |

Now multiply last two tables to obtain TF-IDF for each term in each document:

| | Sentence 1 | Sentence 2 | Sentence 3 |
| - | - | - | - |
| good | 0.5 * 0 = 0 | 0 | 0 |
| boy | 0.5 * 0.176 = 0.088 | 0 * 0.176 = 0 | 1/3 * 0.176 = 0.0587 |
| girl | 0 * 0.176 = 0 | 0.5 * 0.176 = 0.088 | 1/3 * 0.176 = 0.0587 |

From the results above, we can see that for sentence 1, there is some semantic meaning for the word `boy` due to its relatively high value, and in sentence 2 for the word `girl`.

Advantages:
- Better for information retrieval and keyword extraction in documents;
- Balances word importance: considers how often a term appears in the document and in the whole corpus, balancing out common and rare words, thus performing a count normalisation; 
- Simple and easy: the algorithm is quite easy to understand; it's like an improved version of BoW;
- Widely used: it is a well-known method that is widely used

Disadvantages:
- Semantic information is not considered: it doesn't consider word order and the meaning of words;
- TF-IDF gives importance to uncommon / rare words: if a word is very rare, its IDF score will be very high, which can make a term seem more important than it really is; 
- Length bias: this algorithm can be biased towards longer documents, as a word is likely to appear more times in a long document than a short one.

Note about training TF-IDF:
- On training corpus, you calculate TF and IDF, and TF-IDF
- Subsequently, as part of trained TF-IDF vectoriser, you store the values for IDF (as they are corpus-specific), but NOT TF (as it's calculated per each document)
- Later, when transforming textual data as per the trained TF-IDF vectoriser, you re-use IDF values from training, but recalculate the new TF values from the new document



**Variations on calculating TF**

- Binary: 1 if word appears, 0 otherwise
- Normalize the count (sometimes this is the default)

$$tf(t, d) = \cfrac{count(t, d)}{\sum\limits_{t' \in terms(d)} count(t', d)}$$

- Take the log (reduces the influence of extreme values)

$$tf(t, d) = log(1 + count(t, d))$$

**Variations on calculating IDF**

- Trivial IDF: we just set IDF to 1, in which case TF-IDF = TF (aka bag-of-words)

- Smooth IDF: prevents us from ever getting a value of 0

$$idf(t) = log \cfrac{N}{N(t) + 1} + 1$$

- IDF Max: use max frequency of term

$$idf(t) = log \cfrac{max _{t' \in terms(d)} N(t')}{N(t)}$$

- Probabilistic IDF: gives us log-odds (logits)

$$idf(t) = log \cfrac{N - N(t)}{N(t)}$$

````python
"""
TF-IDF by hand
M1a
"""
import numpy as np
import pandas as pd
import math

sentences = [
    'good boy',
    'good girl',
    'boy girl good'
]

### Construct a vocabulary
vocabulary = list(set( ' '.join(sentences).split(' ') ))
print(f"> Vocabulary: \n{vocabulary}")

### Calculate TF
TF = []
for i in sentences:
    i_tokenised = i.split(' ')
    i_counts = {i:i_tokenised.count(i) for i in vocabulary}
    TF_sub = []
    i_len = len(i_tokenised)
    for j in vocabulary:
        TF_sub.append( i_counts[j] / i_len )
    TF.append(TF_sub)

TF_array = np.array(TF)
print(f"\n> TF: \n{TF_array}")
TF_df = pd.DataFrame(TF_array, columns=vocabulary)
for i in TF_df.columns:
    TF_df[i] = TF_df[i].round(1)
TF_df['index'] = TF_df.index + 1
TF_df['index'] = 'Sentence ' + TF_df['index'].astype(str)
TF_df = TF_df.set_index('index')
display( TF_df )

### Calculate IDF
print('> IDF')
IDF = []
corpus_len = len(sentences)
for i in vocabulary:
    a = len([ j for j in sentences if i in j ])
    IDF.append( math.log(corpus_len / a, 10) )
IDF_array = np.array(IDF)
display(IDF_array)

### Calculate TF-IDF vector
print('> TF-IDF')
TF_IDF_array = TF_array * IDF_array
display(TF_IDF_array)
````

````python
"""
TF-IDF by hand
````
````python
M1c
"""
import numpy as np
import pandas as pd
import math

sentences = [
    'good boy',
    'good girl',
    'boy girl good'
]

# Construct a vocabulary
vocabulary = sorted(
    list(set( ' '.join(sentences).split(' ') ))
)
print(f"> Vocabulary: \n{vocabulary}")

def text_processing(x: list[str]):
    return [i.split(' ') for i in sentences]

sentences_vector = text_processing(sentences)
print(f"""
Vectorised sentences: 
{sentences_vector}""")

# Let's learn IDF from the vectorised sentences
def fit_idf(documents: list[list], 
            vocabulary: list[str]
            ) -> np.array:
    """
    Returns learned IDF matrix from a list of documents.
    """
    IDF = np.array(
        [
            math.log(
                len(documents) / len([j for j in documents if i in j]),
                10
            ) 
            for i in vocabulary
        ]
    )
    return IDF

idf = fit_idf(sentences_vector, vocabulary)
print(f"""
IDF: 
{idf}
shape {idf.shape}""")

def calculate_tf(documents: list[list],
                 vocabulary: list
                 ) -> np.array:
    """
    For a list of given sentences, calculate np.array of TF values.
    """
    TF = np.array(
        [
            [i.count(j) / len(i) for j in vocabulary] for i in documents
        ]
    )
    return TF
# calculate_tf(sentences_vector, vocabulary)

def calculate_tf_idf(documents: list[list],
                     learned_idf: np.array,
                     vocabulary: list
                     ) -> np.array:
    """
    Calculate TF-IDF.
    """
    tf = calculate_tf(documents, vocabulary)
    tf_idf = tf * idf
    return tf_idf

tf_idf = calculate_tf_idf(sentences_vector,
                          idf,
                          vocabulary
                          )

print(f"""
TF-IDF:
{tf_idf}
shape {tf_idf.shape}""")

# Get the most important term for each sentence
important_terms_indices = [i.argmax() for i in tf_idf]
print(important_terms_indices)
important_terms_words = [vocabulary[i] for i in important_terms_indices]
important_terms_tf_idf_scores = [tf_idf[i][important_terms_indices[i]] for i in range(len(important_terms_indices))]
print("\nMost important terms:")
for i, j in zip(important_terms_words, important_terms_tf_idf_scores):
    print(f"token: {i} , tf-idf score: {j:.4f}")
````

## Embeddings

Embeddings:
- a technique to get numeric representation of data
- Techniques to **capture similarities between words / semantic meaning of words** by first automatically extracting features from words and sentences. 
- Addresses the limitations of previous approaches by ensuring that similar words have similar vectors, and that arrays are dense, not sparse. 

Word embedding is one of the most important techniques in natural language processing(NLP), where words are mapped to vectors of real numbers. Word embedding is capable of capturing the meaning of a word in a document, semantic and syntactic similarity, relation with other words. It also has been widely used for recommender systems and text classification. For e.g. honest and good are similar words, but older methods cannot capture relationship between these two words.

It is similar to the classical ML classification tasks, where you have features for each data points. Each word is represented by an embedding (fixed size vector), which is essentially a vector consisting of features representing each word (e.g. `is_person?`, `is_country?`, `is_event?`, etc, that are created automatically by an algorithm). In these vectors, words that have similar meaning or context usage are stored close to each other; therefore, two words are simliar if their word vectors are similar. 


For example, if I were to come up with word features by hand, I could come up with something like this: 



Of course, in real-world applications these features (embeddings) are not crafted by hand; instead, they are learned by NN during training. 

For instance, a NN can be trained to predict the next word in a sentence:
- We train a NN to predict the next word in a phrase
- By training, it assigns weights in the first hidden layer, which essentially assigns numbers to each word in different dimensions, and thus it learns to group them based on meaning - generates embeddings;
- this results in similar words having similar high-dimensional embeddings; 
- Finally, having similar words with similar embeddings means training a NN to process language is easier, because learning how one word is used helps learn how similar words are used.

> A general rule of thumb is to set the number of dimensions in your embedding vectors equal to the 4th root of the vocabulary size. For example, if your vocabulary consists of 10000 words, the rule says your embedding vectors should have 10 dimensions. However, similar to the batch size or number of hidden layers in a neural network, you should try different embedding sizes to figure out which is best for the task at hand.

Word embeddings can be trained using various algorithms: 
- Techniques based on CBOW, Skip-gram: 
  - Word2Vec, 
  - GloVe, 
  - fastText
- Techniques based on transformer architecture: 
  - BERT, 
  - GPT

### Word2Vec

Word2Vec is a NN-based CBOW and Skip-gram architectures, better at capturing semantic information. It finds similarities among words by using cosine similarity metric. Offers two NN-based variants - CBOW and Skip-gram. 

Captures local context of words. During training, it only considers neighboring words to capture the context.

````python
def find_analogies(w1, 
                   w2, 
                   w3, 
                   verbose = True
                   ):
    """
    w1 - w2 + w3 = ?
    also can be seen as:
    w1 - w2 = ? - w3

    For example,
    w1 = king, w2 = man, w3 = woman
    king - man + woman = ? 
    or 
    king - man = ? - woman
    """
    r = word_vectors.most_similar(
        positive = [w1, w3], 
        negative = [w2]
    )
    print(f"{w1} - {w2} = {r[0][0]} - {w3}")
    if verbose:
        print(f" > {r}")
    return None

find_analogies('king', 'man', 'woman')
find_analogies('france', 'paris', 'london')
find_analogies('france', 'paris', 'rome')
find_analogies('paris', 'france', 'italy') # poor result for some reason for this one
find_analogies('japan', 'japanese', 'chinese')
find_analogies('december', 'november', 'june', 'september')
find_analogies('miami', 'florida', 'texas')
find_analogies('einstein', 'scientist', 'painter')
find_analogies('man', 'woman', 'she')
````

# Language models

> Essentially, language models take as input a sentence / sequence of words and predict the next word (so you could say there are as many classes as there are words in the corpus).

A language model uses machine learning to conduct a probability distribution over words. Language models learn from text and can be used for producing original text, predicting the next word in a text, speech recognition, optical character recognition and handwriting recognition.

It is a self-supervised learning because we don't need to label the data - the next word itself is the label. 


# LLM

A large language model (LLM) is a language model consisting of a neural network with many parameters (typically billions of weights or more), trained on large quantities of unlabeled text using self-supervised learning or semi-supervised learning. It is considered to be Generative AI. 



Two types of LLM:
- Base LLM
  - Predicts next word, based on text training data
  - Ex1: enter "My friend is", output is the prompt extended with more words: "My friend is ugly".
  - Ex2: enter "What is the capital of England?", output - paraphrasings: "England's capital?", "What is the population of England?"
- Instruction tuned LLM
  - Tries to follow instructions. 
  - Often fine-tuned with RLHF (reinforcement learning with human feedback)
  - Ex1: enter "What is the capital of England?", output - "The capital of England is London."

## ChatGPT

ChatGPT limitations:
- Hallucinations: The LLM can make statements that sound plausible but are not true. For instance, if you ask it about a product which doesn't exist of a real company, it would still produce an output;
  - Can be reduced by: first find relevant information, then answer the question based on the relevnt information

Guidelines for prompting:

**Principles**:
1) Write clear and specific instructions (not necessarily short).
2) Give the model time and opportunity to "think"

There are also some tactics.

**Tactic 1: use delimiters to clearly indicate distinct parts of the input**

- Delimiters: `"""`, `<>`, `<tag></tag>`, `:`
- Delimiters are a useful technique to avoid Prompt Injection

````python
nlp = spacy.load('en_core_web_sm')
````

````python
def extract_countries(text):
    doc = nlp(text)
    # first, extract GPE / countries from the text
    GPEs = [ent.text for ent in doc.ents if ent.label_ == 'GPE']
    # standardise countries' names to ISO3
    ISO3 = coco.convert(names=GPEs, to='ISO3')
    # if only one country is found in GPEs, coco.convert will produce string instead of list, so let's account for that
    if type(ISO3) == str:
        ISO3 = [ISO3]
    # remove duplicates
    ISO3_set = list(set(ISO3))
    return ' | '.join(ISO3_set)

a1 = extract_countries("""
                  Different countries: 
                  Russia, 
                  the Russian Federation,
                  USA, 
                  The United States of America, 
                  The United States, 
                  The US
""")
print(a1)

a2 = extract_countries("""
                       In this text there is no information on countries
""")
print(a2)

extract_countries("""
                  Events in Russia - a general overview
                  """)
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import re
from sklearn.pipeline import Pipeline
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
````

````python
df = pd.read_csv('example_datasets/NLP/IMDB_Dataset.csv')
df.head()
````

````python
def text_clean(text):
    text = text.lower()
    text = re.sub("""[,.?:'"]""", '', text)
    text = re.sub('<br />|<br>|</br>', '', text)
    return text

df['proc'] = df['review'].apply(lambda x: text_clean(x))
df.head()
````

````python
import json
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

vocab_size = 10000
embedding_dim = 16
max_length = 100
trunc_type='post'
padding_type='post'
oov_tok = "<OOV>"
training_size = 20000
````

````python
# !wget --no-check-certificate \
#     https://storage.googleapis.com/learning-datasets/sarcasm.json \
#     -O /tmp/sarcasm.json
````

````python
with open('example_datasets/sarcasm.json', 'r') as f:
    datastore = json.load(f)

sentences, labels, urls = [], [], []
for item in datastore:
    sentences.append(item['headline'])
    labels.append(item['is_sarcastic'])
    urls.append(item['article_link'])

print(len(sentences))
print(sentences[:5])
print(labels[:5])
print(urls[:5])
````

````python
"""
train-test split
"""
training_sentences = sentences[0:training_size]
testing_sentences = sentences[training_size:]
training_labels = labels[0:training_size]
testing_labels = labels[training_size:]
````

````python
tokenizer = Tokenizer(num_words = vocab_size, oov_token = '<OOV>')
tokenizer.fit_on_texts(training_sentences)
word_index = tokenizer.word_index
word_index
````

````python
training_sequences = tokenizer.texts_to_sequences(training_sentences)
training_padded = pad_sequences(
    training_sequences, 
    maxlen = max_length,
    padding = padding_type, 
    truncating = trunc_type
)
training_padded = np.array(training_padded)
training_labels = np.array(training_labels)

testing_sequences = tokenizer.texts_to_sequences(testing_sentences)
testing_padded = pad_sequences(
    testing_sequences, 
    maxlen = max_length,
    padding = padding_type, 
    truncating = trunc_type
)
testing_padded = np.array(testing_padded)
testing_labels = np.array(testing_labels)

print(training_sequences)
print(training_padded)
````

````python
"""
model 1
"""
model_1 = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length = max_length), ### number of parameters: 16 * 10,000 = 160,000
    tf.keras.layers.GlobalAveragePooling1D(), ### converts 160,000 parameters into 16 parameters; explanation of GlobalAveragePooling1D: https://stackoverflow.com/questions/75067335/what-does-globalaveragepooling1d-do-in-keras
    tf.keras.layers.Dense(24, activation = 'relu'), ### number of parameters: 16*24 + 24 = 408
    tf.keras.layers.Dense(1, activation = 'sigmoid') ### number of parameters: 24*1 + 1 = 25
])
model_1.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
display( model_1.summary() )

num_epochs = 30
history = model_1.fit(training_padded, training_labels, epochs = num_epochs, validation_data = (testing_padded, testing_labels), verbose = 2)

````

````python
"""
model 2

- added two dropout layers
"""
model_2 = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length = max_length),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(24, activation = 'relu'),
    tf.keras.layers.Dense(1, activation = 'sigmoid')
])
model_2.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
display( model_2.summary() )

num_epochs = 30
history = model_2.fit(training_padded, training_labels, epochs = num_epochs, validation_data = (testing_padded, testing_labels), verbose = 2)

````

## Generating text

````python
"""
LSTM - generating text
example from https://www.youtube.com/watch?v=ZMudJXhsUpY&list=WL&index=3
"""

import tensorflow as tf

from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding, LSTM, Dense, Bidirectional
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
import numpy as np 
````

````python
tokenizer = Tokenizer()

data = open('example_datasets/irish-lyrics-eof.txt').read()

corpus = data.lower().split("\n")

tokenizer.fit_on_texts(corpus)
total_words = len(tokenizer.word_index) + 1

print(tokenizer.word_index)
print(total_words)
````

````python
input_sequences = []
for line in corpus:
	token_list = tokenizer.texts_to_sequences([line])[0]
	for i in range(1, len(token_list)):
		n_gram_sequence = token_list[:i+1]
		input_sequences.append(n_gram_sequence)

# pad sequences 
max_sequence_len = max([len(x) for x in input_sequences])
input_sequences = np.array(pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre'))

# create predictors and label
xs, labels = input_sequences[:,:-1],input_sequences[:,-1]

ys = tf.keras.utils.to_categorical(labels, num_classes=total_words)
````

````python
print(tokenizer.word_index['in'])
print(tokenizer.word_index['the'])
print(tokenizer.word_index['town'])
print(tokenizer.word_index['of'])
print(tokenizer.word_index['athy'])
print(tokenizer.word_index['one'])
print(tokenizer.word_index['jeremy'])
print(tokenizer.word_index['lanigan'])
````

````python
model = Sequential()
model.add(Embedding(total_words, 100, input_length=max_sequence_len-1))
model.add(Bidirectional(LSTM(150)))
model.add(Dense(total_words, activation='softmax'))
adam = Adam(lr=0.01)
model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
#earlystop = EarlyStopping(monitor='val_loss', min_delta=0, patience=5, verbose=0, mode='auto')
history = model.fit(xs, ys, epochs=100, verbose=1)
#print model.summary()
print(model)
````
